```python
import os
import docx2txt
import docx
from datasets import Dataset
import zipfile

# Function to read .docx files
def read_docx(file_path):
    doc = docx.Document(file_path)
    text = "\n".join([para.text for para in doc.paragraphs])
    return text if text.strip() else None  # Return None if the content is empty

# Function to read .doc files using docx2txt
def read_doc(file_path):
    if file_path.endswith(".doc"):
        try:
            # Try to open the file as a zip file
            with zipfile.ZipFile(file_path) as zipf:
                # Check if the expected file exists in the archive
                if 'word/document.xml' in zipf.namelist():
                    text = docx2txt.process(file_path)
                    return text if text.strip() else None  # Return None if the content is empty
                else:
                    print(f"Skipping file {file_path} as it does not contain word/document.xml.")
                    return None  # Return None for invalid files
        except zipfile.BadZipFile:
            print(f"Skipping file {file_path} as it is not a valid .doc file.")
            return None  # Return None for invalid files
    else:
        print(f"Skipping file {file_path} as it is not a .doc file.")
        return None  # Return None for invalid files

# Directory containing the speeches
speeches_dir = 'speeches'

# List to hold the text data
speeches = []

# Iterate through the files in the directory
for filename in os.listdir(speeches_dir):
    file_path = os.path.join(speeches_dir, filename)

    # Process .docx files
    if filename.endswith(".docx"):
        speech = read_docx(file_path)
        if speech:
            speeches.append(speech)

    # Process .doc files
    elif filename.endswith(".doc"):
        speech = read_doc(file_path)
        if speech:
            speeches.append(speech)

# Print the number of speeches collected
print(f"Number of speeches collected: {len(speeches)}")

# Print the first 5 speeches
print("\nFirst 5 speeches:\n")
for i, speech in enumerate(speeches[:10]):
    print(f"Speech {i + 1}:")
    print(speech)
    print("\n" + "-"*50 + "\n")

# Combine all speeches into a single dataset
data = {"text": speeches}

# Create Hugging Face dataset
dataset = Dataset.from_dict(data)

# Display a sample of the dataset
print(dataset[0])
```

```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel, Trainer, TrainingArguments

# Load pre-trained GPT-2 tokenizer and model
model_name = "gpt2"
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)


# Add a padding token to the tokenizer
tokenizer.pad_token = tokenizer.eos_token

# Tokenize the dataset
def tokenize_function(examples):
    # Tokenize input text
    encodings = tokenizer(examples["text"], truncation=True, padding="max_length", max_length=512)

    # Create labels which are the same as the input text
    labels = encodings["input_ids"].copy()

    # Mask padding tokens in labels
    labels = [-100 if token == tokenizer.pad_token_id else token for token in labels]

    encodings["labels"] = labels
    return encodings

# Tokenize the dataset created from .docx and .doc files
tokenized_dataset = dataset.map(tokenize_function, batched=True)

# Define training arguments
training_args = TrainingArguments(
    output_dir='results',            # Directory where results and checkpoints will be saved
    evaluation_strategy="epoch",       # Evaluate the model after every epoch
    per_device_train_batch_size=2,     # Batch size for training
    per_device_eval_batch_size=2,      # Batch size for evaluation
    num_train_epochs=3,                # Number of training epochs
    weight_decay=0.01,                 # Weight decay to reduce overfitting
)

# Initialize the Trainer
trainer = Trainer(
    model=model,                       # The GPT-2 model
    args=training_args,                # Training arguments
    train_dataset=tokenized_dataset,   # The tokenized dataset for training
    eval_dataset=tokenized_dataset    # The tokenized dataset for evaluation
)

# Fine-tune the model
trainer.train()

model.save_pretrained('results')
tokenizer.save_pretrained('results')
```

```python
import os
import docx2txt
import docx
import zipfile

# Function to read .docx files
def read_docx(file_path):
    doc = docx.Document(file_path)
    text = "\n".join([para.text for para in doc.paragraphs])
    return text if text.strip() else None  # Return None if the content is empty

# Function to read .doc files using docx2txt
def read_doc(file_path):
    if file_path.endswith(".doc"):
        try:
            # Try to open the file as a zip file
            with zipfile.ZipFile(file_path) as zipf:
                # Check if the expected file exists in the archive
                if 'word/document.xml' in zipf.namelist():
                    text = docx2txt.process(file_path)
                    return text if text.strip() else None  # Return None if the content is empty
                else:
                    print(f"Skipping file {file_path} as it does not contain word/document.xml.")
                    return None  # Return None for invalid files
        except zipfile.BadZipFile:
            print(f"Skipping file {file_path} as it is not a valid .doc file.")
            return None  # Return None for invalid files
    else:
        print(f"Skipping file {file_path} as it is not a .doc file.")
        return None  # Return None for invalid files

# Directory containing the speeches
speeches_dir = 'speeches'

# Dictionary to hold the speech contents with file names as keys

speeches_dict = {}

# Iterate through the files in the directory
for filename in os.listdir(speeches_dir):
    file_path = os.path.join(speeches_dir, filename)

    # Process .docx files
    if filename.endswith(".docx"):
        speech_content = read_docx(file_path)
        if speech_content:
            # Store content with the file name (without extension) as the key
            speeches_dict[os.path.splitext(filename)[0]] = speech_content

    # Process .doc files
    elif filename.endswith(".doc"):
        speech_content = read_doc(file_path)
        if speech_content:
            # Store content with the file name (without extension) as the key
            speeches_dict[os.path.splitext(filename)[0]] = speech_content

# The resulting speeches_dict now contains file names as keys and speech content as values
# Example: {'Speech1': 'Speech content...', 'Speech2': 'Another speech content...', ...}

# Print the first 5 speech contents (shortened to 50 characters)
print("First 5 speeches (50 characters each):")
for i, (title, content) in enumerate(speeches_dict.items()):
    if i < 5:  # Print the first 5 entries
        print(f"{i+1}. {title}: {content[:50]}...")
    else:
        break
```
